```python
import numpy as np
import matplotlib.pyplot as plt
!wget -nc https://raw.githubusercontent.com/NYU-IEP-2022-2023-Assignments/QuickBite5/main/manhattan_sales_under_10k_sqft.txt

sqft, price = np.loadtxt('manhattan_sales_under_10k_sqft.txt', skiprows = 1, unpack=True)
price = price/1e6;
```

```python
plt.plot(sqft,price, 'b.')
plt.xlabel('square feet')
plt.ylabel('sale price ($million)')
plt.title('manhattan real estate sales under 10,000 square feet');
```

Above, I've plotted the reported sale price vs. square footage for manhattan property sales in the last 12 months. I got this data [from nyc](https://www.nyc.gov/site/finance/taxes/property-rolling-sales-data.page) and cleaned it up a little, but it still has some problems. 

One of these problems is that deed transfers without compensation are reported as $0 sales. I'd like to get rid of those

`sqft` contains the square footage

`price` contains the price in megadollars

Please remake the plot above, but only plot data points for which `price > 0`

hint: logical indexing is your friend!


```python
PriceGreaterThanZero = price[price > 0]
sqft2 = sqft[price > 0]

plt.plot(sqft2,PriceGreaterThanZero, 'b.')
plt.xlabel('square feet')
plt.ylabel('sale price ($million)')
plt.title('manhattan real estate sales under 10,000 square feet');

'''
I just wanted to see what exactly was being removed so I highlighted the excluded values in green on the same plot.
pricezero = price[price == 0]
sqft3 = sqft[price == 0]
plt.plot(sqft3, pricezero, 'g.')
plt.show()
'''

''' just wanted to see what they'd look like right next to each other

plt.plot(sqft,price, 'r.')
plt.xlabel('square feet')
plt.ylabel('sale price ($million)')
plt.title('manhattan real estate sales under 10,000 square feet');
plt.show()''' 
```
